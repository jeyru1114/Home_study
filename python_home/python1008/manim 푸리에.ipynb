```python
import numpy as np
import matplotlib.pyplot as plt

# Define the frequencies and amplitudes
frequencies = [5, 30, 50]
amplitudes = [1, 0.5, 2]  # Adjust the amplitudes as needed

# Function to get a signal for a given frequency and amplitude
def get_signal(amplitude, frequency):
    x = np.linspace(0, 100, 1000)
    signal = amplitude * np.sin(2 * np.pi * frequency * x / 100)
    return signal

# Combine the signals
combined_signal = np.sum([get_signal(a, f) for a, f in zip(amplitudes, frequencies)], axis=0)

# Create a time array
time = np.linspace(0, 100, len(combined_signal))

# Calculate the Fourier transform of the combined signal
fourier_transform = np.fft.fft(combined_signal)
frequencies_fft = np.fft.fftfreq(len(time), d=time[1]-time[0])

# Plot the combined signal in the time domain
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(time, combined_signal, color='b', label='Combined Signal (Time Domain)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Plot the Fourier transform of the combined signal
plt.subplot(2, 1, 2)
plt.plot(frequencies_fft, np.abs(fourier_transform), color='r', label='Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 2)  # Limit the x-axis to positive frequencies
plt.legend()

plt.tight_layout()
plt.show()
```
